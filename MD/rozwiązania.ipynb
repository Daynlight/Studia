## Zadanie 1.A.1

```python
def fun_many(data):
    a = len(data)
    count = 0
    listFinal = []
    for  i in range(a):
        ist = [0]*3
        ist[0] = data[i]
        for j in range(a):
            ist[1] = data[j]
            for k in range(a):
                ist[2] = data[k]
                listFinal.append(list(ist))
                count+=1
    print(listFinal)
    print(count)

def fun_single(data):
    a =len(data)
    count = 0
    listFinal = []
    for i in range (a):
        ist = [0] * 3
        ist[0] = data[i]
        for j in range(a):
            if data[j] != ist[0]:
                ist[1] = data[j]
                for k in range(a):
                    if data[k] != ist[0] and data[k] != ist[1]:
                        ist[2] = data[k]
                        listFinal.append(list(ist))
                        count+=1
                        
    print(listFinal)
    print(count)
```

```python
data = 'Ala, Basia, Celina'
data = sorted(data.split(", "))
fun_many(data)
fun_single(data)
```

## Zadanie 1.A.2

```python
import itertools

def losowanie_kul(n):
    kule = list(range(1, n + 1))
    permutacje = list(itertools.permutations(kule, 4))

    # Konwersja krotek na listy
    permutacje = [list(p) for p in permutacje]

    # Warunki
    war1 = [p for p in permutacje if p[0] == 3]
    war2 = [p for p in permutacje if p[0] % 2 == 0]
    war3 = [p for p in permutacje if p[0] % 2 == 0 or p[0] == 3]
    war4 = [p for p in permutacje if p[0] % 2 == 0 or p[0] % 3 == 0]

    # Wyświetlenie wyników
    for lista in [war1, war2, war3, war4]:
        print(lista)
        print(len(lista))
```

```python
n = 4
losowanie_kul(n)
```

## Zadanie 1.A.3

```python
import itertools

def cukiernia(n):
    # 1. Każdy deser innego rodzaju (kombinacje bez powtórzeń)
    wariant1 = list(itertools.combinations(range(1, n + 1), 3))
    wariant1 = [list(p) for p in wariant1]

    # 2. Z powtórzeniami, ale bez znaczenia kolejności (multizbiory)
    wariant2 = list(itertools.combinations_with_replacement(range(1, n + 1), 3))
    wariant2 = [list(p) for p in wariant2]

    # Wynik
    print(wariant1)
    print(len(wariant1))
    print(wariant2)
    print(len(wariant2))
```

```python
# Przykład użycia
n = 4
cukiernia(n)
```

## Zadanie 1.A.4

```python
import itertools

def the_office_corrected(n):
    osoby = list(range(1, n + 1))

    # --- Wariant 1: 3 + 2 + 1 ---
    konfiguracje_1 = []
    for trojka in itertools.combinations(osoby, 3):
        pozostali_po_3 = [x for x in osoby if x not in trojka]
        for dwojka in itertools.combinations(pozostali_po_3, 2):
            pozostaly = [x for x in pozostali_po_3 if x not in dwojka]
            for jedynka in itertools.combinations(pozostaly, 1):
                konfiguracje_1.append([list(trojka), list(dwojka), list(jedynka)])

    # --- Wariant 2: trzy rozróżnialne pary (projekty) ---
    konfiguracje_2 = []

    for szostka in itertools.combinations(osoby, 6):
        osoby_szostka = list(szostka)
        # Wszystkie możliwe permutacje osób
        for perm in itertools.permutations(osoby_szostka):
            para1 = tuple(sorted(perm[0:2]))
            para2 = tuple(sorted(perm[2:4]))
            para3 = tuple(sorted(perm[4:6]))
            projekt = [para1, para2, para3]
            konfiguracje_2.append(tuple(projekt))

    # Usuwamy powtórzenia (te same 3 pary mogą wystąpić w różnych permutacjach)
    konfiguracje_2_unikalne = set(konfiguracje_2)

    print(len(konfiguracje_1))
    print(len(konfiguracje_2_unikalne))
```

```python
# Test
n = 6
the_office_corrected(n)
```

## Zadanie 1.A.5

```python
import itertools

def rozwiazania_suma(n, warunek):
    wyniki = []
    for x in itertools.product(range(n + 1), repeat=5):
        if sum(x) == n and warunek(x):
            wyniki.append(list(x))
    return wyniki

def rozwiazania_min1(n):
    return rozwiazania_suma(n - 5, lambda x: all(xi >= 0 for xi in x))

def zadanie1_A_5(n):
  wyniki_a5a = rozwiazania_suma(n, lambda x: True)
  print(wyniki_a5a)
  print("Liczba rozwiązań (a):", len(wyniki_a5a))

  wyniki_a5b = rozwiazania_suma(n, lambda x: x[4] >= 3)
  print(wyniki_a5b)
  print("Liczba rozwiązań (b):", len(wyniki_a5b))


  wyniki_a5c = [[xi + 1 for xi in w] for w in rozwiazania_min1(n)]
  print(wyniki_a5c)
  print("Liczba rozwiązań (c):", len(wyniki_a5c))

  wyniki_a5d = rozwiazania_suma(n, lambda x: x[2] < 3)
  print(wyniki_a5d)
  print("Liczba rozwiązań (d):", len(wyniki_a5d))
```

```python
zadanie1_A_5(5)
zadanie1_A_5(11)
```

## Zadanie 2.A.4

```python
import itertools

def konfiguracje_zdjec_zgrupowani(podgrupa, pozostali):
    wszystkie_konfiguracje = []
    # Permutacje wewnętrzne podgrupy
    for podgrupa_perm in itertools.permutations(podgrupa):
        paczka = list(podgrupa_perm)
        # Pozostałe osoby jako pojedyncze elementy
        skladniki = [paczka] + [[p] for p in pozostali]
        # Wszystkie permutacje bloków (paczka + reszta)
        for perm in itertools.permutations(skladniki):
            uklad = []
            for fragment in perm:
                uklad.extend(fragment)
            wszystkie_konfiguracje.append("".join(uklad))
    return sorted(set(wszystkie_konfiguracje))  # unikalne i posortowane
```

## Zadanie 2.A.5

```python
from more_itertools import distinct_permutations

def sposoby_wydania_kwoty(n):
    suma = n * 10
    kombinacje = []
    max_20 = suma // 20
    for i in range(max_20 + 1):
        reszta = suma - i * 20
        if reszta % 50 == 0:
            j = reszta // 50
            kombinacje.append([20]*i + [50]*j)
    # Wszystkie unikalne permutacje dla każdej kombinacji
    wyniki = []
    for k in kombinacje:
        wyniki.extend(distinct_permutations(k))
    return list(wyniki)

def wypisz(l, descending=False):
    string_list = []
    for el in l:
        string = "".join(str(int(a / 10)) for a in el)
        string_list.append(string)
    # Sort ascending or descending based on flag
    string_list.sort(reverse=descending)
    for s in string_list:
        print(s)
```

```python
# A.5.a - n = 12 (kwota = 120 PLN)
A5a = sposoby_wydania_kwoty(12)
# A.5.b - n = 15 (kwota = 150 PLN)
A5b = sposoby_wydania_kwoty(15)

wypisz(A5a)
wypisz(A5b)
```
