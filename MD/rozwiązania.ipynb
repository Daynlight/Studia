```python
def fun_many(data):
    a = len(data)
    count = 0
    listFinal = []
    for  i in range(a):
        ist = [0]*3
        ist[0] = data[i]
        for j in range(a):
            ist[1] = data[j]
            for k in range(a):
                ist[2] = data[k]
                listFinal.append(list(ist))
                count+=1
    print(listFinal)
    print(count)

def fun_single(data):
    a =len(data)
    count = 0
    listFinal = []
    for i in range (a):
        ist = [0] * 3
        ist[0] = data[i]
        for j in range(a):
            if data[j] != ist[0]:
                ist[1] = data[j]
                for k in range(a):
                    if data[k] != ist[0] and data[k] != ist[1]:
                        ist[2] = data[k]
                        listFinal.append(list(ist))
                        count+=1
                        
    print(listFinal)
    print(count)
```

```python
import itertools

def losowanie_kul(n):
    kule = list(range(1, n + 1))
    permutacje = list(itertools.permutations(kule, 4))

    # Konwersja krotek na listy
    permutacje = [list(p) for p in permutacje]

    # Warunki
    war1 = [p for p in permutacje if p[0] == 3]
    war2 = [p for p in permutacje if p[0] % 2 == 0]
    war3 = [p for p in permutacje if p[0] % 2 == 0 or p[0] == 3]
    war4 = [p for p in permutacje if p[0] % 2 == 0 or p[0] % 3 == 0]

    # Wyświetlenie wyników
    for lista in [war1, war2, war3, war4]:
        print(lista)
        print(len(lista))
```

```python
import itertools

def cukiernia(n):
    # 1. Każdy deser innego rodzaju (kombinacje bez powtórzeń)
    wariant1 = list(itertools.combinations(range(1, n + 1), 3))
    wariant1 = [list(p) for p in wariant1]

    # 2. Z powtórzeniami, ale bez znaczenia kolejności (multizbiory)
    wariant2 = list(itertools.combinations_with_replacement(range(1, n + 1), 3))
    wariant2 = [list(p) for p in wariant2]

    # Wynik
    print(wariant1)
    print(len(wariant1))
    print(wariant2)
    print(len(wariant2))
```

## Zadanie 1.A.4

```python
import itertools

def the_office_corrected(n):
    osoby = list(range(1, n + 1))

    # --- Wariant 1: 3 + 2 + 1 ---
    konfiguracje_1 = []
    for trojka in itertools.combinations(osoby, 3):
        pozostali_po_3 = [x for x in osoby if x not in trojka]
        for dwojka in itertools.combinations(pozostali_po_3, 2):
            pozostaly = [x for x in pozostali_po_3 if x not in dwojka]
            for jedynka in itertools.combinations(pozostaly, 1):
                konfiguracje_1.append([list(trojka), list(dwojka), list(jedynka)])

    # --- Wariant 2: trzy rozróżnialne pary (projekty) ---
    konfiguracje_2 = []

    for szostka in itertools.combinations(osoby, 6):
        osoby_szostka = list(szostka)
        # Wszystkie możliwe permutacje osób
        for perm in itertools.permutations(osoby_szostka):
            para1 = tuple(sorted(perm[0:2]))
            para2 = tuple(sorted(perm[2:4]))
            para3 = tuple(sorted(perm[4:6]))
            projekt = [para1, para2, para3]
            konfiguracje_2.append(tuple(projekt))

    # Usuwamy powtórzenia (te same 3 pary mogą wystąpić w różnych permutacjach)
    konfiguracje_2_unikalne = set(konfiguracje_2)

    print(len(konfiguracje_1))
    print(len(konfiguracje_2_unikalne))
```

## Zadanie 2.A.4

```python
import itertools

def konfiguracje_zdjec_zgrupowani(podgrupa, pozostali):
    wszystkie_konfiguracje = []
    # Permutacje wewnętrzne podgrupy
    for podgrupa_perm in itertools.permutations(podgrupa):
        paczka = list(podgrupa_perm)
        # Pozostałe osoby jako pojedyncze elementy
        skladniki = [paczka] + [[p] for p in pozostali]
        # Wszystkie permutacje bloków (paczka + reszta)
        for perm in itertools.permutations(skladniki):
            uklad = []
            for fragment in perm:
                uklad.extend(fragment)
            wszystkie_konfiguracje.append("".join(uklad))
    return sorted(set(wszystkie_konfiguracje))  # unikalne i posortowane
```

```python
from more_itertools import distinct_permutations

def sposoby_wydania_kwoty(n):
    suma = n * 10
    kombinacje = []
    max_20 = suma // 20
    for i in range(max_20 + 1):
        reszta = suma - i * 20
        if reszta % 50 == 0:
            j = reszta // 50
            kombinacje.append([20]*i + [50]*j)
    # Wszystkie unikalne permutacje dla każdej kombinacji
    wyniki = []
    for k in kombinacje:
        wyniki.extend(distinct_permutations(k))
    return list(wyniki)

def wypisz(l, descending=False):
    string_list = []
    for el in l:
        string = "".join(str(int(a / 10)) for a in el)
        string_list.append(string)
    # Sort ascending or descending based on flag
    string_list.sort(reverse=descending)
    for s in string_list:
        print(s)
```

```python
# A.5.a - n = 12 (kwota = 120 PLN)
A5a = sposoby_wydania_kwoty(12)
# A.5.b - n = 15 (kwota = 150 PLN)
A5b = sposoby_wydania_kwoty(15)

wypisz(A5a)
wypisz(A5b)
```

```python
from itertools import combinations

def algorytm1(n, m, l):
    osoby = list(range(n))
    wszyscy = list(combinations(osoby, l))
    tylko_mezczyzni = list(combinations(range(m), l))
    
    # Usuń drużyny złożone wyłącznie z mężczyzn
    z_kobieta = [team for team in wszyscy if team not in tylko_mezczyzni]
    return z_kobieta

def algorytm2(n, m, l):
    drużyny = []
    for k in range(m, n):  # Kobiety: indeksy m do n-1
        # Kobieta na pozycji k
        pozostali = list(range(k))  # osoby o niższym indeksie
        for team in combinations(pozostali, l - 1):
            drużyny.append(tuple(sorted(team + (k,))))
    return drużyny
```

```python
# n, m, l = map(int, input().split(", "))  # np. 6, 2, 2
n,m,l = 6, 2, 2
# o = input()  # np. Adam, Bartek, Celina, Daria, Ewa, Tosia
o = "Adam, Bartek, Celina, Daria, Ewa, Tosia"

# Imiona jako lista
osoby = [x.strip() for x in o.split(",")]

# Wygeneruj drużyny
druzyny1 = algorytm1(n, m, l)
druzyny2 = algorytm2(n, m, l)

# Zamień indeksy na imiona i posortuj imiona w drużynach
druzyny1_imiona = [sorted([osoby[i] for i in team]) for team in druzyny1]
druzyny2_imiona = [sorted([osoby[i] for i in team]) for team in druzyny2]

# Posortuj całą listę drużyn alfabetycznie
druzyny1_imiona.sort()
druzyny2_imiona.sort()

# Wypisz wyniki
print(druzyny1_imiona)
print(len(druzyny1_imiona))
print(druzyny2_imiona)
print(len(druzyny2_imiona))
```

## Zadanie 3.A.2

```python
from itertools import combinations

# Własna funkcja: wszystkie dwuczęściowe podziały zbioru
def bipartitions(s):
    s = list(s)
    n = len(s)
    wynik = []
    for i in range(1 << n):
        part1 = [s[j] for j in range(n) if (i >> j) & 1]
        part2 = [s[j] for j in range(n) if not (i >> j) & 1]
        wynik.append((sorted(part1), sorted(part2)))
    return wynik

def algorytm1(n, k):
    kule = list(range(1, n + 1))
    wynik = []
    for malowane in combinations(kule, k):
        for N, C in bipartitions(malowane):
            # wykluczamy tylko sytuację, gdy oba są puste
            if len(N) == 0 and len(C) == 0:
                continue
            B = sorted(set(kule) - set(N) - set(C))
            wynik.append([N, C, B])
    return sorted(wynik)

def algorytm2(n, k):
    kule = list(range(1, n + 1))
    wynik = []
    for j in range(k + 1):
        for N in combinations(kule, j):
            pozostale = list(set(kule) - set(N))
            for C in combinations(pozostale, k - j):
                B = sorted(set(kule) - set(N) - set(C))
                wynik.append([sorted(N), sorted(C), B])
    return sorted(wynik)
```

## Zadanie 3.A.3

```python
def set_partitions(iterable, k=None):
    """
    Yield the set partitions of *iterable* into *k* parts. Set partitions are
    not order-preserving.

    >>> iterable = 'abc'
    >>> for part in set_partitions(iterable, 2):
    ...     print([''.join(p) for p in part])
    ['a', 'bc']
    ['ab', 'c']
    ['b', 'ac']


    If *k* is not given, every set partition is generated.

    >>> iterable = 'abc'
    >>> for part in set_partitions(iterable):
    ...     print([''.join(p) for p in part])
    ['abc']
    ['a', 'bc']
    ['ab', 'c']
    ['b', 'ac']
    ['a', 'b', 'c']

    """
    L = list(iterable)
    n = len(L)
    if k is not None:
        if k < 1:
            raise ValueError(
                "Can't partition in a negative or zero number of groups"
            )
        elif k > n:
            return

    def set_partitions_helper(L, k):
        n = len(L)
        if k == 1:
            yield [L]
        elif n == k:
            yield [[s] for s in L]
        else:
            e, *M = L
            for p in set_partitions_helper(M, k - 1):
                yield [[e], *p]
            for p in set_partitions_helper(M, k):
                for i in range(len(p)):
                    yield p[:i] + [[e] + p[i]] + p[i + 1 :]

    if k is None:
        for k in range(1, n + 1):
            yield from set_partitions_helper(L, k)
    else:
        yield from set_partitions_helper(L, k)


# tutaj wpisz swój kod 
from itertools import combinations, permutations, product

def partitions_rest(rest):
    if not rest:
        yield ([], [], [])
    else:
        for assignment in product(range(3), repeat=len(rest)):
            U1, U2, U3 = [], [], []
            for ball, urn_idx in zip(rest, assignment):
                if urn_idx == 0:
                    U1.append(ball)
                elif urn_idx == 1:
                    U2.append(ball)
                else:
                    U3.append(ball)
            yield (U1, U2, U3)

def algorytm1(n):
    kule = list(range(1, n+1))
    wynik = []

    for perm in permutations(kule[:3]):
        pozostale = kule[3:]
        for podzial in partitions_rest(pozostale):
            U1 = [perm[0]] + podzial[0]
            U2 = [perm[1]] + podzial[1]
            U3 = [perm[2]] + podzial[2]
            wynik.append((tuple(sorted(U1)), tuple(sorted(U2)), tuple(sorted(U3))))

    wynik = list(set(wynik))
    wynik.sort()

    return [ [list(u1), list(u2), list(u3)] for (u1,u2,u3) in wynik]

def algorytm2(n):
    kule = list(range(1, n+1))
    wynik = []

    for triple in combinations(kule, 3):
        rest = [x for x in kule if x not in triple]
        for perm in permutations(triple):
            if not rest:
                U1, U2, U3 = [perm[0]], [perm[1]], [perm[2]]
                wynik.append((tuple(sorted(U1)), tuple(sorted(U2)), tuple(sorted(U3))))
            else:
                for assignment in product(range(3), repeat=len(rest)):
                    U1, U2, U3 = [perm[0]], [perm[1]], [perm[2]]
                    for ball, urn_idx in zip(rest, assignment):
                        if urn_idx == 0:
                            U1.append(ball)
                        elif urn_idx == 1:
                            U2.append(ball)
                        else:
                            U3.append(ball)
                    wynik.append((tuple(sorted(U1)), tuple(sorted(U2)), tuple(sorted(U3))))
    return wynik

def compositions(n, k):
    """
    Generate all compositions of integer n into k positive integers.
    """
    if k == 1:
        yield (n,)
    else:
        for i in range(1, n - k + 2):  # i ≥ 1
            for comp in compositions(n - i, k - 1):
                yield (i,) + comp

def ordered_partitions(lst, sizes):
    """
    Generate all ordered partitions of list lst into subsets of sizes.
    """
    if len(sizes) == 1:
        if len(lst) == sizes[0]:
            yield [lst]
        return

    first_size = sizes[0]
    for subset in combinations(lst, first_size):
        rest = [x for x in lst if x not in subset]
        for tail in ordered_partitions(rest, sizes[1:]):
            yield [list(subset)] + tail

def algorytm3(n):
    kule = list(range(1, n + 1))
    wynik = set()

    for comp in compositions(n, 3):
        for partition in ordered_partitions(kule, comp):
            # Sort elements within each group, but keep the group order
            normalized = tuple(tuple(sorted(group)) for group in partition)
            wynik.add(normalized)

    wynik = sorted(wynik)  # Sort for consistent output
    return [ [list(u) for u in triple] for triple in wynik ]

def algorytm4(n):
    kule = list(range(1, n + 1))
    wynik = set()

    for assignment in product(range(3), repeat=n):
        U1, U2, U3 = [], [], []
        for ball, urn_idx in zip(kule, assignment):
            if urn_idx == 0:
                U1.append(ball)
            elif urn_idx == 1:
                U2.append(ball)
            else:
                U3.append(ball)
        # Skip any partition with an empty subset
        if U1 and U2 and U3:
            partition = (tuple(sorted(U1)), tuple(sorted(U2)), tuple(sorted(U3)))
            wynik.add(partition)

    wynik = sorted(wynik)
    return [[list(u1), list(u2), list(u3)] for (u1, u2, u3) in wynik]
```

```python
def a(n, c, a_values):
    r = len(c)
    if n < r:
        return a_values[n]
    return sum(c[i] * a(n - i - 1, c, a_values) for i in range(r))
```
